```python
diadora = []
diadora_product = []
for i in range(1, 17):
    diadora.append("http://www.diadora-tw.com.tw/prod.asp?Pages=" + str(i))
for i in range(1, 9):
    diadora.append("http://www.diadora-tw.com.tw/prod.asp?id=2&Pages=" + str(i))
for i in range(1, 4):
    diadora.append("http://www.diadora-tw.com.tw/prod.asp?id=595&Pages=" + str(i))
for i in range(1, 3):
    diadora.append("http://www.diadora-tw.com.tw/prod.asp?id=596&Pages=" + str(i))
for i in range(1, 6):
    diadora.append("http://www.diadora-tw.com.tw/prod.asp?id=405&Pages=" + str(i))
```

```python
for href in diadora:
    driver.get(href)
    bodyElement = driver.find_element_by_tag_name("body")
    body_text = driver.execute_script("return arguments[0].innerHTML;", bodyElement)
    soup = BeautifulSoup(body_text, "html.parser")
    
    for r in soup.findAll("div", "pd"):
        for a in r.findAll("a"):
            diadora_product.append("http://www.diadora-tw.com.tw/" + a["href"])
```

```python
wb = Workbook()
ws = wb.active
ws.append(["產品", "品牌", "型號", "客群", "價格", "類別", "介紹"])

for href in diadora_product:
    driver.get(href)
    bodyElement = driver.find_element_by_tag_name("body")
    body_text = driver.execute_script("return arguments[0].innerHTML;", bodyElement)
    soup = BeautifulSoup(body_text, "html.parser")
    
    product = soup.find("div", "pdtop").find("h1").text
    product = product.replace("/","")
    brand = "diadora"
    model = soup.findAll("div", "itemtxt")[0].text
    user = soup.findAll("div", "itemtxt")[1].text
    user = product[-1:] if user == "經典復古鞋" else user[:1]
    price = soup.find("div", "itemtxt2").text[4:]
    category = ""
    category = [i for i in shoes_category if i in product]
    category = "".join(category)
    category = "經典鞋" if category == "" else category
    introduction = soup.find("div", "summary").text.strip()
    introduction = introduction.replace("‧","")
    print("產品: " + product)
    print("品牌: " + brand)
    print("型號: " + model)
    print("客群: " + user)
    print("價格: " + price)
    print("類別: " + category)
    print("介紹: " + introduction)
    
    ws.append([product, brand, model, user, price, category, introduction])

wb.save('C:/Users/pro/Desktop/運動/diadora.xls')
driver.close()
```
