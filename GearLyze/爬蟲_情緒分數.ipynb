```python
import csv
import requests
from bs4 import BeautifulSoup
import jieba
import jieba.posseg as pseg
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
import time
```

```python
jieba.set_dictionary("C:/Users/pro/jieba/dict.txt.big")
```

```python
#去除停用詞、斷詞、找出名詞動詞
stop_words = []
for i in open("C:/Users/pro/Desktop/專案/dictionary/stop_words.txt", encoding = "utf-8"):
    stop_words.append(str(i))
    
def sent2word(sentence, stop_words = stop_words):
    words = pseg.cut(sentence)
    word = []
    symbol = ". —、，~～,.《》<>/\[]{}()（）｢「」$！!？?#。：:；;\n"
    for w in words:
        if w.word not in " ".join(stop_words) and w.word not in symbol:
            if w.flag[0].upper() == "N" or w.flag[0].upper() == "V":
                word.append(w.word)
    return word
```

```python
#判斷極性
negative=[]
for i in open("C:/Users/pro/Desktop/專案/dictionary/ntusd/ntusd-negative.txt", encoding = "utf-8"):
    negative.append(str(i))

positive=[]
for i in open("C:/Users/pro/Desktop/專案/dictionary/ntusd/ntusd-positive.txt", encoding = "utf-8"):
    positive.append(str(i))


def findpolarity(term):
    point=0
    for a in range(len(positive)):
        if(term.strip() == positive[a].strip()):
            point=1
    
    for b in range(len(negative)):
        if(term.strip() == negative[b].strip()):
            point=-1
    return point
```

```python
#將網址文章儲存csv檔(情緒分數使用)_鉅亨網
urlstore = open("C:/Users/pro/Desktop/爬蟲/鉅亨網網址_1.txt", encoding = "utf-8")

f = open("C:/Users/pro/Desktop/爬蟲/情緒分數_鉅亨網.csv", "w", encoding = "utf-8", newline="")
writer = csv.writer(f)
writer.writerow(["日期", "標題", "情緒分數"])

# driver = webdriver.Chrome(ChromeDriverManager().install())

for url in urlstore:
    driver.get(url)
    
    time.sleep(5)
    bodyElement = driver.find_element_by_tag_name("body")
    body_text = driver.execute_script("return arguments[0].innerHTML;", bodyElement)
    soup = BeautifulSoup(body_text, "html.parser")
    
    time.sleep(3)
    date = soup.find("time")
    date = date.text[:10].replace("/", "")
    title = soup.find("h1")
    title = title.text.strip()
    paragraph = soup.findAll("p")
    
    fraction = 0

    for p in paragraph:
        word = sent2word(p.text.replace("\n", ""))
        for w in word:
#             print(w, findpolarity(w))
            fraction += findpolarity(w)
            
    data = [date, title, fraction]
    print(data)
    writer.writerow(data)

urlstore.close()
f.close()
```

```python
#將txt檔文章儲存csv檔(情緒分數使用)_公開資料觀測站
import os, sys

company = ["2330",
           "2317",
           "3105",
           "3037",
           "2327",
           "6488",
           "2492",
           "6462",
           "3443",
           "5483"]

for c in company:
    f = open("C:/Users/pro/Desktop/爬蟲/情緒分數_" + c + ".csv", "w", newline="")
    writer = csv.writer(f)
    writer.writerow(["日期", "標題", "情緒分數"])
    
    path = "C:/Users/pro/Desktop/爬蟲/公開資料觀測站/" + c
    files = os.listdir(path) #得到文件夾下的所有文件名稱
    for file in files:
        f_1 = open(path + "/" + file , "r", encoding = "utf-8")
        news = f_1.read()
        f_1.close()

        word = sent2word(news)
        fraction = 0
        for w in word:
            fraction += findpolarity(w)
        
        data = [file[:8], file[9:file.find(".")], fraction]
        print(data)
        writer.writerow(data)

    f.close()
```
